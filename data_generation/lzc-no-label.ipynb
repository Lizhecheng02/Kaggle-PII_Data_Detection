```python
from openai import OpenAI
from tqdm import tqdm
import os
import random
import pandas as pd
import numpy as np
```

```python
os.environ["OPENAI_API_KEY"] = ""
```

```python
def get_response(msg):
    messages = [
        {
            "role": "user",
            "content": msg
        }
    ]

    temperature = random.uniform(0.6, 0.8)
    client = OpenAI()
    response = client.chat.completions.create(
        model="gpt-3.5-turbo-0125",
        messages=messages,
        temperature=temperature
    )

    output = response.choices[0].message.content
    return output
```

```python
df = pd.DataFrame(columns=['document', 'full_text', 'tokens', 'trailing_whitespace', 'labels'])
```

```python
from spacy.lang.en import English

en_tokenizer = English().tokenizer


def tokenize_with_spacy(text, tokenizer=en_tokenizer):
    tokenized_text = tokenizer(text)
    tokens = [token.text for token in tokenized_text]
    trailing_whitespace = [bool(token.whitespace_) for token in tokenized_text]
    return {"tokens": tokens, "trailing_whitespace": trailing_whitespace}
```

```python
num = 200

for i in tqdm(range(num), desc="Generating no label essays"):
    prompt = """
Write an essay in first person that details your experience of applying a specific tool or approach to address a complex challenge. Every time you write this essay, you need to think of a new experience, use your imagination.
Here are some rules you must follow:
- Do not show your personal information in the essay.
- The essay can contain emails, street addresses, usernames, IDs, or websites that are not relevant to you or other students, they can be related with instructors or companies.
- You should cite at least one article, state its content and author. The article you cite could be fabricated by yourself.
- The essay should not only narrate the process but also critically analyze the effectiveness of the chosen tool or approach, reflecting on its strengths and potential limitations.
- The essay should have a title and try to creat a insterting title.
- The comma and period in the essay must be followed by a space.
- The essay should contain at least 1000 words.
Now, return your essay only.
    """
    print(prompt)

#     prompt = """
# Select one of the four design thinking tools presented in the course (listed below) that you are interested in applying to a challenge of your choice. Your completed reflection should be written in English and be between five and ten paragraphs in length.

# Tools: Visualization, Storytelling, Mind Mapping or Learning Launch.

# Elements:
# - Describe your challenge, including all relevant information.
# - In your own words, briefly describe the tool you selected (e.g., what it is and why you selected it for your challenge -- including any appropriate video lecture references).
# - Describe how you applied the tool you selected to your challenge (e.g., what you did and how the tool was applied effectively or ineffectively).
# - Describe the insight you gained from applying the tool you selected to your challenge (e.g., how an insight affected your thinking about the challenge and about design thinking more broadly).
# - Describe what you might do differently next time -- applying the same tool you selected or a different one and the reason(s) why.

# Here are some rules you must follow:
# - Do not show your personal information in the essay.
# - The essay can contain emails, street addresses, usernames, IDs, or websites that are not relevant to you or other students, they can be related with instructors or companies.
# - You should cite at least one article, state its content and author. The article you cite could be fabricated by yourself.
# - The essay should have a title and try to creat a insterting title.
# - The essay should contain at least one thousand words.
#     """

    document = str(i)
    full_text = get_response(prompt)
    print(full_text)
    output_dict = tokenize_with_spacy(text=full_text)
    tokens = output_dict['tokens']
    trailing_whitespace = output_dict["trailing_whitespace"]
    labels = []
    for i in range(len(tokens)):
        labels.append("O")

    print(len(tokens), len(trailing_whitespace), len(labels))

    new_row = pd.DataFrame({
        'document': [document],
        'full_text': [full_text],
        'tokens': [tokens],
        'trailing_whitespace': [trailing_whitespace],
        'labels': [labels]
    })
    print(new_row.dtypes)
    df = pd.concat([df, new_row], ignore_index=True)
    df.to_json("no_label_data.json", orient="records")
```
