```python
import pandas as pd
```

```python
origin_train = pd.read_json("./kaggle_dataset/competition/train.json")
origin_train.head()
```

```python
origin_train.columns
```

```python
type(origin_train["trailing_whitespace"][0][0])
```

```python
moth_0120 = pd.read_csv("./kaggle_dataset/pii_dataset_moth_0120.csv")
moth_0120.head()
```

```python
moth_0120.columns
```

```python
moth_0120.rename(columns={"text": "full_text"}, inplace=True)
```

```python
origin_moth_0120 = pd.concat([origin_train, moth_0120], axis=0, ignore_index=True)
origin_moth_0120["document"] = list(range(len(origin_moth_0120)))
origin_moth_0120.reset_index(drop=True, inplace=True)
origin_moth_0120.head()
```

```python
type(origin_moth_0120["trailing_whitespace"][0][0])
```

```python
origin_moth_0120.shape
```

```python
origin_moth_0120.dropna(inplace=True)
origin_moth_0120.reset_index(drop=True, inplace=True)
origin_moth_0120.shape
```

```python
def func(x):
    if isinstance(x, list):
        return x
    elif isinstance(x, str):
        return eval(x)
```

```python
origin_moth_0120["tokens"] = origin_moth_0120["tokens"].apply(func)
origin_moth_0120["labels"] = origin_moth_0120["labels"].apply(func)
origin_moth_0120["trailing_whitespace"] = origin_moth_0120["trailing_whitespace"].apply(func)
```

```python
origin_moth_0120.tail()
```

```python
origin_moth_0120.to_json("./combined_dataset/origin_moth_0120.json", orient="records")
```
